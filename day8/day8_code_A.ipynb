```python
text_input = open('input.txt', 'r').read().splitlines()

#pseudorules

"""
If nop, move to the next elem
If acc, add the arg to the accumulator var and move to the next elem,
If jmp, jump the arg given, forward or backwards, using relative index position

Need an accumlator var, a 'visited' var to ensure we don't get stuck and a while loop.

""";
```

```python
visited_indexes = []
accumulator = 0
position = 0
safety_net = 0 #to make sure we don't do something stupid

while len(visited_indexes) == len(set(visited_indexes)) and safety_net < 10000:
    
    #Get the instruction and arg
    instruction, argument = text_input[position].split()
    
    print(f'1. The position is {position}')
    print(f'2. The accumulator is currently {accumulator}')
    print(f'3. Performing {instruction} with argument {argument} \n\n')
    
    #Evaulate instruction and update position and accumulator as required
    
    if instruction == 'nop':
        visited_indexes.append(position)
        position += 1
    elif instruction == 'acc':
        visited_indexes.append(position)
        accumulator += int(argument)
        position += 1
    elif instruction == 'jmp':
        visited_indexes.append(position)
        position += int(argument)

    safety_net += 1
    
print(f'Loop finished and the accumulator prior to the loop breaking was {accumulator - int(argument)}')
```
